```python
import pacsys
from pacsys import Device, ArrayDevice, ScalarDevice
```

```python
# set global defaults
pacsys.configure(role="testing")
```

```python
read_tmp = Device("M:OUTTMP")
read_array = ArrayDevice("B:IRMS06[0:10]")
output1 = ScalarDevice("Z:ACLTST")  # this helps type checkers by enforcing float type, optional
output2 = Device("Z:CUBE")
```

```python
read_tmp.read()
```

```python
read_tmp.get()
```

```python
read_array.read()
```

```python
z1 = output1.read()
```

```python
output1.write(z1 + 1.0)
```

```python
assert z1 + 1.0 == output1.read()
```

```python
def cb(value, handle):
    print(f"Value changed to {value}")


handle = output1.with_event("p,1000").subscribe(cb)
```

```python
handle.stop()
```
